# Excluding points with ejecta contamination

```python
## Check data filtering

lr0, lx0, lr, dlr, delta_radio, lx, dlx_l, dlx_u, delta_xrays, source_names, unique_names, unique_D, unique_D_prob, t = get_data_arrays(names = ["GX 339-4"], interp=True, rerun = False, save=False, incl_Fr_uplims=True, type_source="BH")
delta = delta_radio.astype(int) # Convert boolean to int (1 for detection, 0 for upper limit)


# x:
lx_scaled = lx/lx0
log_lx_scaled = np.log10(lx_scaled) # x
log_dlx_l_scaled = np.log10(lx / (lx - dlx_l)) # = log(lx/lx0) - log( (lx - dlx_l)/ lx0 ) 
log_dlx_u_scaled = np.log10( ( lx + dlx_u) / lx ) # = log( (lx + dlx_u)/ lx0 ) - log(lx/lx0)
log_dlx_scaled = np.maximum(log_dlx_l_scaled , log_dlx_u_scaled )

# y:
lr_scaled = lr/lr0
log_lr_scaled = np.log10(lr_scaled) # y
log_dlr_l_scaled = np.log10(lr / (lr - dlr))
log_dlr_u_scaled = np.log10( ( lr + dlr) / lr )
log_dlr_scaled = np.maximum(log_dlr_l_scaled , log_dlr_u_scaled )

x = log_lx_scaled
y = log_lr_scaled
dx = log_dlx_scaled
dy = log_dlr_scaled


## Filter for region of interest

t0 = 58964
t1 = 59083
gx339_mask = (lx > 2.7e34) & ( (t<t0) | (t>t1) )
lr_filtered, lx_filtered, dlr_filtered, dlx_l_filtered, dlx_u_filtered ,  delta_filtered = lr[gx339_mask], lx[gx339_mask], dlr[gx339_mask], dlx_l[gx339_mask], dlx_u[gx339_mask],  delta[gx339_mask]
x_filtered, y_filtered, dx_filtered, dy_filtered, delta_filtered = x[gx339_mask], y[gx339_mask], dx[gx339_mask], dy[gx339_mask], delta[gx339_mask]


## Plot the selected data points

plt.errorbar(lx, lr, yerr=dlr, xerr=[dlx_l, dlx_u], fmt=".")
plt.errorbar(lx_filtered, lr_filtered, yerr=dlr_filtered, xerr=[dlx_l_filtered, dlx_u_filtered], fmt=".", color="red")
plt.xscale("log")
plt.yscale("log")
#plt.xlim(min_Lx, max_Lx)
#plt.ylim(min_Lr, max_Lr_2)
plt.show()
```

```python
## Look to see how distance affects GX339

plt.errorbar(x, y, yerr=dy, xerr=dx, fmt=".")
plt.errorbar(x_filtered, y_filtered, yerr=dy_filtered, xerr=dx_filtered, fmt=".", color="red")
plt.errorbar(x_filtered + 2*np.log10(8 / 10), y_filtered + 2*np.log10(8 / 10), yerr=dy_filtered, xerr=dx_filtered, fmt=".", color="green")
plt.errorbar(x_filtered + 2*np.log10(12/ 10), y_filtered+ 2*np.log10(12 / 10), yerr=dy_filtered, xerr=dx_filtered, fmt=".", color="pink")
#plt.xlim(min_Lx, max_Lx)
#plt.ylim(min_Lr, max_Lr_2)
plt.show()
```

```python
# Compare with ODR result
# Should be quite similar since the intrinsic scatter is small
# Recall though that this only deals with detections

import scipy.odr as odr

mask_filtered_det = delta_filtered.astype(bool)
x_filtered_det = x_filtered[mask_filtered_det]
y_filtered_det = y_filtered[mask_filtered_det]
dx_filtered_det = dx_filtered[mask_filtered_det]
dy_filtered_det = dy_filtered[mask_filtered_det]

# Define the linear model for ODR
def linear_func(beta, x):
    return beta[0] * x + beta[1]

# Create a model object
model = odr.Model(linear_func)

# Create a data object with errors
data = odr.RealData(x_filtered_det, y_filtered_det, sx=dx_filtered_det, sy= dy_filtered_det)

# Set up the ODR with an initial guess
odr_instance = odr.ODR(data, model, beta0=[1.0, 0.0])

# Run the regression
output = odr_instance.run()

# Get the fitted parameters
slope, intercept = output.beta
slope_err, intercept_err = output.sd_beta
print(f"Fitted parameters:")
print(f"  Slope = {slope:.3f}  {slope_err:.3f}")
print(f"  Intercept = {intercept:.3f}  {intercept_err:.3f}")
```
